```python
import pickle as pk
A = pk.load(open('global.pk','rb'))

# sample nodes
N = ['L4','L34','L79','L123']

import random

def gen_hex_colour_code():
   return '#'+''.join([random.choice('0123456789ABCDEF') for x in range(6)])

C = {}
```

```python
l = len(A)
D = {}
for i in range(l):
    D[i] = {}
    D[i]['n1'] = A[i]['n1']
    D[i]['n2'] = A[i]['n2']
    tmp = A[i]['envelope']
    ce = A[i]['envelope'][0][1]
    cex = A[i]['envelope'][0][0]
    D[i]['x'] = [xx[0]-cex for xx in tmp]
    D[i]['y'] = [xx[1]-ce for xx in tmp]
    D[i]['z'] = [xx-ce for xx in A[i]['z']]
    D[i]['slope'] = A[i]['slope']
    D[i]['r2'] = A[i]['r2']
```

# Translation by largest envelope point

```python
p = []
C = {}
for roi in N:
    tit = 'SOURCE '+roi
    s1 = figure(plot_width=400, plot_height=400, title = tit,y_axis_label = "Probability(log scale)",x_axis_label = "Distance")
    datax = [xx['x'] for xx in D.values() if xx['n1'] == roi]
    datay = [xx['y'] for xx in D.values() if xx['n1'] == roi]
    fx = [item for sublist in datax for item in sublist]
    fy = [item for sublist in datay for item in sublist]
    F = lslinear(fx,fy)
    datap = [(xx['x'],xx['y'],xx['n2']) for xx in D.values() if xx['n1'] == roi]
    for w in datap:
        try:
            color = C[w[2]]
        except KeyError:
            color = gen_hex_colour_code()
            C[w[2]] = color
        s1.circle(w[0],w[1],color=color)
    tmp = str(-F['slope']/100)
    lg = 'R2 = '+str(num.ceil(F['r2']))
    s1.line(F['x'],F['z'],color='red',legend =lg )
    p.append(s1)
    tit = 'TARGET '+roi
    s2 = figure(plot_width=400, plot_height=400, title = tit,y_axis_label = "Probability(log scale)",x_axis_label = "Distance")
    datax = [xx['x'] for xx in D.values() if xx['n2'] == roi]
    datay = [xx['y'] for xx in D.values() if xx['n2'] == roi]
    datap = [(xx['x'],xx['y'],xx['n1']) for xx in D.values() if xx['n2'] == roi]
    for w in datap:
        try:
            color = C[w[2]]
        except KeyError:
            color = gen_hex_colour_code()
            C[w[2]] = color
        s2.circle(w[0],w[1],color=color)
    fx = [item for sublist in datax for item in sublist]
    fy = [item for sublist in datay for item in sublist]
    F = lslinear(fx,fy)
    s2.line(F['x'],F['z'],color='red')
    s2.line(F['x'],F['z'],color='red',legend = 'R2 = '+str(num.ceil(F['r2'])))
    p.append(s2)

grid = gridplot([[p[0],p[1]], [p[2],p[3]], [p[4],p[5]], [p[6],p[7]]])
show(grid)
```

```python
i = 0
ff = lambda x: x+ np.random.normal(-2, 1, len(x))
ffx = lambda x: x+ np.random.normal(0, 10, len(x))
gt = {}
gt['r2'] = []
gt['sl'] = []
gt['_r2'] = []
gt['_sl'] = []
for w in A:
    if w['r2']<90: continue
    i = i +1
    tmp  = w['envelope']
    a = [xx[0] for xx in tmp]
    a = ff(a)
    b = [xx[1] for xx in tmp]
    b = ff(b)
    gt['r2'].append(w['r2'])
    gt['sl'].append(-w['slope']/100)
    D = lslinear(a,b)
    gt['_r2'].append(D['r2'])
    gt['_sl'].append(D['slope'])
```
