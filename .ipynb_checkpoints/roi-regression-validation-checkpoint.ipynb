```python
import pickle as pk
A = pk.load(open('local.pk','rb'))


import random

def gen_hex_colour_code():
   return '#'+''.join([random.choice('0123456789ABCDEF') for x in range(6)])

C = {}
```

```python
l = len(A)
D = {}
for i in range(l):
    D[i] = {}
    D[i]['n1'] = A[i]['n1']
    D[i]['n2'] = A[i]['n2']
    tmp = A[i]['envelope']
    ce = num.sum([xx[1] for xx in A[i]['envelope']])/len(A[i]['envelope'])
    cex = num.sum([xx[0] for xx in A[i]['envelope']])/len(A[i]['envelope'])
    D[i]['x'] = [xx[0]-cex for xx in tmp]
    D[i]['y'] = [xx[1]-ce for xx in tmp]
    D[i]['z'] = [xx-ce for xx in A[i]['z']]
    D[i]['slope'] = A[i]['slope']
    D[i]['r2'] = A[i]['r2']
p = []
C = {}
N = ['L'+str(i+1) for i in range(180)]
R = {}
for roi in N:
    # source taget
    datax = [xx['x'] for xx in D.values() if xx['n1'] == roi]
    datay = [xx['y'] for xx in D.values() if xx['n1'] == roi]
    fx = [item for sublist in datax for item in sublist]
    fy = [item for sublist in datay for item in sublist]
    F = lslinear(fx,fy)
    R["s-"+roi]=(F['r2'],F['slope'])
    
    
    # target
    datax = [xx['x'] for xx in D.values() if xx['n2'] == roi]
    datay = [xx['y'] for xx in D.values() if xx['n2'] == roi]
    fx = [item for sublist in datax for item in sublist]
    fy = [item for sublist in datay for item in sublist]
    F = lslinear(fx,fy)
    R["t-"+roi]=(F['r2'],F['slope'],F['std'])
```

```python
def slope_estimate(local,r2,std,s,t):
    S = R["s-"+s]
    T = R["t-"+t]
    # max r2
    fx = [0,0,0]
    model = [(0,0),(0,0),(0,0)]
    r = [r2,S[0],T[0]]
    sl = [local,S[1],T[1]]
    stds = [std,S[-1],T[-1]]
    ind = np.argmax(r)
    fx[0] = sl[ind]
    model[0] = (sl[ind],stds[ind])
    # avg
    fx[1] = np.sum([sl[i]*r[i] for i in range(3)])/np.sum(r)
    model[1] = (fx[1],np.sqrt(np.sum([np.square(stds[i]*r[i]) for i in range(3)])/np.square(np.sum(r))))
    # avg_local
    tmp = np.sum([sl[i]*r[i] for i in range(1,3)])/np.sum(r[1:])
    fx[2] = (sl[0]*r[0]+(100-r[0])*tmp)/100
    a = r[0]/100
    den = np.sum(r[1:])
    b = (r[1]/den)*((100-r[0])/100)
    c = (r[2]/den)*((100-r[0])/100)
    cur = np.sqrt(a**2 * stds[0]**2 + b**2 * stds[1]**2 + c**2 * stds[2]**2)
    model[2] = (fx[2],cur)
    return (fx,model)
```

```python
GT = list(filter(lambda x: x['r2']>=90, A))
g1 = 0
g2 = 0
g3 = 0
G1 = []
G2 = []
S1 = []
S2 = []
S3 = []
for i,XY in enumerate(GT):
    tmp1 = list(zip(XY['a'],XY['b']))
    tmp2 = list(filter(lambda x: x not in XY['envelope'],tmp1))
    x_new = [xx[0] for xx in tmp2]
    y_new = [xx[1] for xx in tmp2]
    D_new = pre_mania2(x_new, y_new)
    s = XY['n1']
    t = XY['n2']
    if (D_new['reason']!='envelope'):
        g1 = g1 + 1
        G1.append([D_new['r2'],D_new['slope'],XY['r2'],-XY['slope']/100])
        fx,model = slope_estimate(D_new['slope'],D_new['r2'],D_new['std'],s,t)
        S1.append([XY['slope'],D_new['slope']]+ fx + model + [(XY['slope'],XY['std'])])
    else:
        g3 = g3 + 1
        fx,model = slope_estimate(0,0,0,s,t)
        S3.append([XY['slope'],0]+ fx + model + [(XY['slope'],XY['std'])])
print("Number of G1 pairs:"+ str(g1))
print("Number of G2 pairs:"+ str(g2))
print("Number of G3 pairs:"+ str(g3))
```

```python
l = len(S1)
model1 = []
model2 = []
model3 = []
for i in range(l):
    tmp = S1[i]
    models = tmp[5:8]
    ax = tmp[-1]
    for q,w in enumerate(models):
        try:
            a,b = overlap(ax[0],w[0],ax[1],w[1])
        except:
            continue
        if q == 0: model1.append((a,b))
        elif q == 1: model2.append((a,b))
        else: model3.append((a,b))
```

```python
p1 = figure(title="Slope Estimation Error",toolbar_location="below",
            background_fill_color="#E8DDCB")

measured = [xx[0] for xx in model1]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p1.line(edges[1:],cdf,legend = 'MAX')

measured = [xx[0] for xx in model2]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p1.line(edges[1:],cdf,line_color="red",legend = 'AVG')

measured = [xx[0] for xx in model3]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p1.line(edges[1:],cdf,line_color="yellow",legend = 'LOCAL-AVG')
p1.legend.location = "bottom_left"

p1.xaxis.axis_label = "Percentage of deviation from ground truth mean"
p1.yaxis.axis_label = "CDF"

p2 = figure(title="Model Similarity",toolbar_location="below",
            background_fill_color="#E8DDCB")

measured = [xx[1] for xx in model1]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p2.line(edges[1:],cdf,legend = 'MAX')

measured = [xx[1] for xx in model2]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p2.line(edges[1:],cdf,line_color="red",legend = 'AVG')

measured = [xx[1] for xx in model3]
hist, edges = np.histogram(measured, density=True, bins=50)
cdf = np.cumsum(hist)
cdf = 100.0*cdf/max(cdf)
p2.line(edges[1:],cdf,line_color="yellow",legend = 'LOCAL-AVG')
p2.legend.location = "bottom_left"
p2.xaxis.axis_label = "Similarity"
p2.yaxis.axis_label = "CDF"
grid = gridplot([[p1, p2]])
show (grid)
```
