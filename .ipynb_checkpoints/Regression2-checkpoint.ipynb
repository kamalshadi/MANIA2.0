# Regressin among all left-hemisphere ROIs for subject 481042

```python
R = roi_regressors()
```

```python
# NEO4J database call for GROUP 1 - Close and Connected
graph = Graph(password="1234")
c0 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE n.hemisphere="L" and r.SUBJECT = 481042
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length as distance ,r._weight as weight, r.SUBJECT as subject
ORDER BY D'''


A = graph.run(c0)
results = []
indices = {}
for i,w in enumerate(A):
    a = np.array(w['distance'])
    b = np.array(w['weight'])
    D = regress_noplot(a,b)
    D['n1'] = w['n1']
    D['n2'] = w['n2']
    D['subject'] = w['subject']
    D['max-weight'] = w['W']
    D['distance'] = w['D']
    results.append(D)
    indices[w['n1']+w['n2']] = i
```

```python
RA = all_regressors(R,results)
```

```python
rois = ['L'+str(i) for i in range(1,181)]
```

```python
val={}
raw={}
types = {'null':0,'adjacent':0,'local':0,'mixed':0, 'r2':0 , 'env':0}
for s in rois:
    for t in rois:
        if s==t:continue

        try:
            ind = indices[s+t]
        except:
            val[s+t] = num.log(1/5000)
            types['null'] = types['null'] + 1
            raw[s+t] = num.log(1/5000)
            continue
        cur = results[ind]
        raw[s+t] = num.log(cur['max-weight']/5000.0)
        reg = RA[s+t]
        if cur['isSuccess']:
            if cur['envelope'][0][0]<1:
                types['adjacent'] = types['adjacent']+1
                val[s+t] = num.log(cur['max-weight']/5000.0)
            else:
                types['local'] = types['local']+1
                pt = cur['envelope'][0]
                val[s+t] = pt[1] - reg[0]*pt[0]
        else:
            types['mixed'] = types['mixed'] + 1
            if cur['reason'] == 'r2':
                types['r2'] = types['r2'] + 1
                pt = cur['envelope'][0]
                if (reg[0]):
                    if reg[0]<0:
                        val[s+t] = pt[1] - reg[0]*pt[0]
                    else:
                        val[s+t] = num.log(cur['max-weight']/5000.0)  
                else:
                    val[s+t] = num.log(cur['max-weight']/5000.0)
            else:
                if len(cur['envelope'])>0:
                    pt = cur['envelope'][0]
                    val[s+t] = pt[1] - reg[0]*pt[0]
                    types['env'] = types['env'] + 1
                else:
                    val[s+t] = num.log(cur['max-weight']/5000.0)
                    types['env'] = types['env'] + 1
```

```python
CC = num.ones((180,180))*num.log(1.0/5000)
CR = num.ones((180,180))*num.log(1.0/5000)
for i in range(1,181):
    for j in range(1,181):
        if i==j:continue
        n1 = 'L'+str(i)
        n2 = 'L'+str(j)
        CC[i-1,j-1] = val[n1+n2]
        CR[i-1,j-1] = raw[n1+n2]
```

```python
from utils import *
```

```python
clip = 1
nar = [0]*(5000-2*clip)
den = [0]*(5000-2*clip)
narc = [0]*(5000-2*clip)
denc = [0]*(5000-2*clip)
T = [0]*(5000-2*clip)
for i in range(clip,5000-clip):
    net = num.zeros((180,180))
    the = num.log(i/5000.0)
    net[CC>the] = 1
    q = i - clip
    nar[q] = NAR(net)
    den[q] = density(net)
    T[q] = the
    net = num.zeros((180,180))
    net[CR>the] = 1
    narc[q] = NAR(net)
    denc[q] = density(net)
```

```python
p1 = figure(title="Density - NAR",toolbar_location="below",
            background_fill_color="#E8DDCB", plot_width=500, plot_height=500)
p1.square(den,nar,color = 'red',legend = 'Corrected')
p1.line(denc,narc,color = 'green',legend = 'Not Corrected')
p1.xaxis.axis_label = 'Density'
p1.yaxis.axis_label = 'NAR'

p2 = figure(title="Threshold - NAR",toolbar_location="below",
            background_fill_color="#E8DDCB", plot_width=500, plot_height=500)
p2.square(T,nar,color = 'red',legend = 'Not Corrected')
p2.line(T,narc,color = 'green',legend = 'Corrected')
p2.xaxis.axis_label = 'Threshold'
p2.yaxis.axis_label = 'NAR'

p3 = figure(title="Threshold - Density",toolbar_location="below",
            background_fill_color="#E8DDCB",plot_width=500, plot_height=500)
p3.square(T,den,color = 'red',legend = 'Corrected')
p3.line(T,denc,color = 'green',legend = 'Not Corrected')
p3.xaxis.axis_label = 'Threshold'
p3.yaxis.axis_label = 'Density'

grid = gridplot([[p1, p2],[p3,None]])
show (grid)
```

# Success rate and failure reasons

```python
T = '''
---------------------------
Category:ZZZ
---------------------------
Number of pairs:AAA
Number of accepted Regressions:BBB
Rejected for R2:CCC
Rejected for Envelope:DDD
'''
cat = ['Weekly connected', 'Somewhat connected', 'strongly connected']
for q,K in enumerate(res):
    R = res[K]
    ZZZ = cat[q]
    AAA = len(R)
    BBB = len([xx for xx in R if xx['isSuccess']])
    CCC = len([xx for xx in R if (~xx['isSuccess'] and xx['reason']=='r2')])
    DDD = len([xx for xx in R if (~xx['isSuccess'] and xx['reason']=='envelope')])
    cur = T.replace('AAA',str(AAA)).replace('BBB',str(BBB)).replace('CCC',str(CCC)).replace('DDD',str(DDD)).replace('ZZZ',ZZZ)
    print(cur)
```
