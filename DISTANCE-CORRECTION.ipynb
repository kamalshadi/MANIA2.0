```python
from py2neo import Node, Relationship, Graph
from numpy.random import rand
import numpy as num
from bokeh.plotting import output_notebook,figure, show
from bokeh.layouts import row,column,gridplot
from bokeh.models import Label
import numpy as np
import matplotlib.pyplot as plt
from curve import *
from bokeh.models import Span
output_notebook()
```

```python
# NEO4J database call for GROUP 1 - Close and Connected
graph = Graph(password="1234")
c1 = '''MATCH (n:ROI)-[r:NOS]->(m:ROI)
WHERE n.hemisphere="L" and r.SUBJECT=481042
RETURN n.name AS n1 ,m.name AS n2,r.length as D,r.weight as W,r._length,r._weight
ORDER BY r.length
limit 8'''

A = graph.run(c1)
# Plotting the data
s1 = []
k1 = []
k2 = []
for i,w in enumerate(A):
    a = w['r._length'] 
    b = w['r._weight']
    k1 = k1 + a
    k2 = k2 + [xx/5000.0 for xx in b]
```

```python
D = {}
l = len(k2)
for i in range(l):
    ind = int(k2[i]*100)
    try:
        D[ind].append(k1[i])
    except KeyError:
        D[ind] = [k1[i]]
```

```python
from scipy.stats import entropy
```

```python
p1 = figure(plot_width=500, plot_height=500, title='Entropy for D(f)',y_axis_label = "Entropy",x_axis_label = "NOS fraction")
p2 = figure(plot_width=500, plot_height=500, title='Mean Distance for D(f)',y_axis_label = "Distance",x_axis_label = "NOS fraction")
p3 = figure(plot_width=500, plot_height=500, title='Std Distance for D(f)',y_axis_label = "Distance",x_axis_label = "NOS fraction")
p4 = figure(plot_width=500, plot_height=500, title='#Samples Distance for D(f)',y_axis_label = "Number",x_axis_label = "NOS fraction")

# Plotting the data
s1 = []
for w in D.keys():
    t= D[w]
    pk, _ = num.histogram(t,bins=40,range=(0,200))
    z = sum(pk)
    pp = [xx/z for xx in pk]
    ent = entropy(pk)
    x_point = w/100
    p1.square(x_point,ent,color='green',size=8)
    if w<0.001:
        continue
    p2.square(x_point,num.mean(t),color='green',size=8)
    p3.square(x_point,num.std(t),color='green',size=8)
    p4.square(x_point,len(t),color='green',size=8)
grid = gridplot([[p1,p2],[p3,p4]])
show (grid)
```
