```python
import pickle as pk
A = pk.load(open('global.pk','rb'))

# sample nodes
N = ['L4','L34','L79','L123']

import random

def gen_hex_colour_code():
   return '#'+''.join([random.choice('0123456789ABCDEF') for x in range(6)])

C = {}
```

# Envelope Scatter Plots

```python
p = []
S1 = {}
S2 = {}
for n in N:
    S1[n] = {}
    env1 = [(xx['envelope'],xx['n2'],xx['slope']) for xx in A if (xx['n1']==n)]
    tit1 = 'Source '+n
    S2[n] = {}
    env2 = [(xx['envelope'],xx['n1'],xx['slope']) for xx in A if (xx['n2']==n)]
    tit2 = 'Target '+n
    s1 = figure(plot_width=400, plot_height=400, title = tit1, y_axis_label = "Probability(log scale)",x_axis_label = "Distance")
    s2 = figure(plot_width=400, plot_height=400, title = tit2, y_axis_label = "Probability(log scale)",x_axis_label = "Distance")
    for target in env1:
        tar = target[0]
        S1[n][target[1]] = target[2]
        try:
            color = C[target[1]]
        except KeyError:
            color = gen_hex_colour_code()
            C[target[1]] = color
#     flat_list = [item for sub in env for item in sub]
        x = [xx[0] for xx in tar if xx[0]>1]
        y = [xx[1] for xx in tar if xx[0]>1]
#     tit = 'Source '+n
        s1.circle(x,y,color=color)
    for source in env2:
        src = source[0]
        S2[n][source[1]] = source[2]
        try:
            color = C[source[1]]
        except KeyError:
            color = gen_hex_colour_code()
            C[source[1]] = color
#     flat_list = [item for sub in env for item in sub]
        x = [xx[0] for xx in src if xx[0]>1]
        y = [xx[1] for xx in src if xx[0]>1]
#     tit = 'Source '+n
        s2.circle(x,y,color=color)
    p.append([s1,s2])
    
grid = gridplot([p[0], p[1], p[2], p[3]])
show (grid)
```

# Slopes

```python
color = ['red','blue','green','black']
from bokeh.models import TickFormatter
rs = [0]*4
rt = [0]*4
txt = ['a']*4
tt = ['a']*4
s = figure(plot_width=600, plot_height=400, y_axis_label = "Slopes to targets/Sources",x_axis_label = "Target/Source")
for i,n1 in enumerate(S1.keys()):
    txt[i] = "Source "+n1
    val = [-S1[n1][xx]/100.0 for xx in S1[n1]]
    rs[i] = s.circle(2*i,val,color=color[i])
for i,n1 in enumerate(S2.keys()):
    tt[i] = "target "+n1
    val = [-S2[n1][xx]/100.0 for xx in S2[n1]]
    rt[i] = s.square(2*i+1,val,color=color[i])
legend = Legend(items=[(txt[i],[rs[i]]) for i in range(4)]+[(tt[i],[rt[i]]) for i in range(4)], location=(1, 30))
s.add_layout(legend, 'right')
s.xaxis.ticker = [0,1, 2, 3,4,5,6,7]
show(s)
```

```python
p = []
S1 = {}
S2 = {}
N = ['L'+str(xx) for xx in range(1,181)]
for n in N:
    S1[n] = [-xx['slope']/100 for xx in A if (xx['n1']==n and xx['r2']>85)]
    S2[n] = [-xx['slope']/100 for xx in A if (xx['n2']==n and xx['r2']>85)]
```

```python
D1 = {}
D2 = {}
from numpy import std
ff = lambda x: std(x)
for n in N:
    if (len(S1[n]) > 0):
        D1[n] = ff(S1[n])
    if (len(S2[n]) > 0):
        D2[n] = ff(S2[n])
```

```python
import numpy
```

```python
import numpy
hist1, edges1 = numpy.histogram(list(D1.values()), density=True, bins=25)
hist2, edges2 = numpy.histogram(list(D2.values()), density=True, bins=25)
```

```python
s = figure(plot_width=600, plot_height=400, y_axis_label = "Histogram",x_axis_label = "Variability")
s.line(edges1,hist1,color= 'green',legend = 'Source Regressors')
s.line(edges2,hist2,color = 'red',legend = 'Target Regressors')
s.legend.location = "center_right"
a = num.median(list(D1.values()))
b = num.median(list(D2.values()))
s.line([a,a],[0,30],color = 'yellow',legend = 'Median (Source Specific)')
s.line([b,b],[0,30],color = 'black',legend = 'Median (Target Specific)')

show(s)
```
